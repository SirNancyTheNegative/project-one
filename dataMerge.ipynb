```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Load data
file_path = "Resources/greenhouse_gas_inventory_data.csv"
gas_df = pd.read_csv(file_path)
gas_df
```

```python
gas_df = gas_df.drop(columns=["category"]).groupby(["country_or_area","year"]).sum()
gas_df
```

```python
# Save data to CSV
file_path = "Resources/EnvTempYearly_Cleaned.csv"
temp_df = pd.read_csv(file_path)
temp_df
```

```python
tempgas_df = temp_df.merge(gas_df,left_on=["Area","Year"],right_on=["country_or_area","year"])
tempgas_df = tempgas_df.rename(columns={"value":"Greenhouse Gases"})
tempgas_df
```

```python
# null values
for column in tempgas_df.columns:
    print(f"Column {column} has {tempgas_df[column].isnull().sum()} null values")
```

```python
# Save data to CSV
file_path = "Resources/Temperature_Greenhouse_merged.csv"
tempgas_df.to_csv(file_path, index=False)
```
