```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Load data
file_path = "Resources/Environment_Temperature_change_clean.csv"
box_df = pd.read_csv(file_path)
box_df
```

```python
year_box_df = box_df.loc[box_df["Months Code"] == 7020].drop(columns=["Months Code","Months"])
year_box_df
```

```python
clean_df = pd.DataFrame(data = year_box_df.values.tolist(),columns=year_box_df.columns)
clean_df
```

```python
# Takes: List of lists
# Returns: Single list with all lists appended piecewise to the first entry of the list of lists
def dimensionshrink(lists):
    oneD = lists[0]
    for entry in lists:
        if entry == lists[0]:
            continue
        for item in entry:
            oneD.append(item)
    return oneD

# Takes: Single list, number of repetitions
# Returns: Single list with each entry of the input repeated that number of times.
def duplicateEntries(initList, numReps):
    endList = []
    for entry in initList:
        for i in range(numReps):
            endList.append(entry)
    return endList

# Takes: Single list, number of repetitions
# Returns: Single list with the input list's contents repeated in order for the number of repetitions.
def duplicateList(initList, numReps):
    endList = []
    for i in range(numReps):
        for entry in initList:
            endList.append(entry)
    return endList

# Takes: Three lists of same length
# Returns: Single list of entries with one entry each of the same three lists
def dataMerger(areas,years,changes):
    endList = []
    for i in range(len(areas)):
        endList.append([areas[i],years[i],changes[i]])
    return endList

# Takes: List of strings
# Returns: List of integers with the first character of the original string removed
def YRemove(strlist):
    endList = []
    for string in strlist:
        endList.append(int(string[1:]))
    return endList
```

```python
# Running the four functions above
change_vals = dimensionshrink(clean_df.drop(columns=["Area"]).values.tolist())
area_vals = duplicateEntries(list(clean_df.Area),len(clean_df.drop(columns=["Area"]).values.tolist()[0]))
year_vals = duplicateList(YRemove(clean_df.drop(columns=["Area"]).columns.tolist()),len(clean_df.Area.values.tolist()))
data = dataMerger(area_vals,year_vals,change_vals)
```

```python
# Compile them into a dataframe
year_df = pd.DataFrame(data=data,columns = ["Area","Year","Temp Change"])
year_df
```

```python
# Save data to CSV
file_path = "Resources/EnvTempYearly_Cleaned.csv"
year_df.to_csv(file_path, index=False)
```
