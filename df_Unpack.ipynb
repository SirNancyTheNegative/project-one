```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Load data
file_path = "Resources/Environment_Temperature_change_clean.csv"
box_df = pd.read_csv(file_path)
box_df
```

```python
year_box_df = box_df.loc[box_df["Months Code"] == 7020].drop(columns=["Months Code","Months"])
year_box_df
```

```python
clean_df = pd.DataFrame(data = year_box_df.values.tolist(),columns=year_box_df.columns)
clean_df
```

```python
# Takes: List of lists
# Returns: Single list with all lists appended piecewise to the first entry of the list of lists
def dimensionshrink(lists):
    oneD = lists[0]
    for entry in lists:
        if entry == lists[0]:
            continue
        for item in entry:
            oneD.append(item)
    return oneD

# Takes: Single list, number of repetitions
# Returns: Single list with each entry of the input repeated that number of times.
def duplicateEntries(initList, numReps):
    endList = []
    for entry in initList:
        for i in range(numReps):
            endList.append(entry)
    return endList

# Takes: Single list, number of repetitions
# Returns: Single list with the input list's contents repeated in order for the number of repetitions.
def duplicateList(initList, numReps):
    endList = []
    for i in range(numReps):
        for entry in initList:
            endList.append(entry)
    return endList

# Takes: Three lists of same length
# Returns: Single list of entries with one entry each of the same three lists
def dataMerger(areas,years,changes):
    endList = []
    for i in range(len(areas)):
        endList.append([areas[i],years[i],changes[i]])
    return endList

# Takes: List of strings
# Returns: List of integers with the first character of the original string removed
def YRemove(strlist):
    endList = []
    for string in strlist:
        endList.append(int(string[1:]))
    return endList

#Takes: A dataframe with columns "country","year","value", and "category"
#Changes the dataframe to add values from "category" as columns and removes "value" and "category" columns
def valSplit(df):
    listlist = []
    for category in df["category"].unique().tolist():
        val_list = []
        
        print("Checking for %s" % category)
        for year in df["year"].unique().tolist():
            for area in df["area"].unique().tolist():
                if area in df.loc[(df.category == category) & (df.year == year)]["area"].values:
                    val_list.append(df.loc[(df.category == category) & (df.year == year) & (df.area == area)]["value"].tolist()[0])
                else:
                    val_list.append(0)
        print("Length of list: %s" % len(val_list))
        #while len(val_list) < len(df.index):
        #    val_list.append(0)
        listlist.append(val_list)
        print("listlist length: %s" % len(listlist))
        
    return listlist

#Takes: two lists
#Returns: A list of lists with the second one appended to the first
def listAppender(list1, list2):
    for item in list2:
        list1.append(item)
    return list1

#Takes: two lists of the same length, and a list of lists with entries of the same length as the first two
#Returns: A list of lists with each entry being the ith entry of each of the inputs as a list.
def dataPrepper(areas, years, values):
    final_data = []
    for i in range(len(areas)):
        entry = []
        entry.append(areas[i])
        entry.append(years[i])
        for item in values:
            entry.append(item[i])
        final_data.append(entry)
    return final_data
```

```python
# Running the four functions above
change_vals = dimensionshrink(clean_df.drop(columns=["Area"]).values.tolist())
area_vals = duplicateEntries(list(clean_df.Area),len(clean_df.drop(columns=["Area"]).values.tolist()[0]))
year_vals = duplicateList(YRemove(clean_df.drop(columns=["Area"]).columns.tolist()),len(clean_df.Area.values.tolist()))
data = dataMerger(area_vals,year_vals,change_vals)
```

```python
# Compile them into a dataframe
year_df = pd.DataFrame(data=data,columns = ["Area","Year","Temp Change"])
year_df
```

```python
# Save data to CSV
file_path = "Resources/EnvTempYearly_Cleaned.csv"
year_df.to_csv(file_path, index=False)
```

```python
# Load data
file_path = "Resources/greenhouse_gas_inventory_data.csv"
gas_df = pd.read_csv(file_path)
gas_df
```

```python
gas_df = gas_df.replace({
    "carbon_dioxide_co2_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent" : "carbon_dioxide",
    'greenhouse_gas_ghgs_emissions_including_indirect_co2_without_lulucf_in_kilotonne_co2_equivalent' : "greenhouse_with_indirect",
    'greenhouse_gas_ghgs_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent' : "greenhouse_without_indirect",
    'hydrofluorocarbons_hfcs_emissions_in_kilotonne_co2_equivalent' : "hydrofluorocarbons",
    'methane_ch4_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent' : "methane",
    'nitrogen_trifluoride_nf3_emissions_in_kilotonne_co2_equivalent' : "nitrogen_trifluoride",
    'nitrous_oxide_n2o_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent' : "nitrous_oxide",
    'perfluorocarbons_pfcs_emissions_in_kilotonne_co2_equivalent' : "perfluorocarbons",
    'sulphur_hexafluoride_sf6_emissions_in_kilotonne_co2_equivalent' : "sulphur_hexafluoride",
    'unspecified_mix_of_hydrofluorocarbons_hfcs_and_perfluorocarbons_pfcs_emissions_in_kilotonne_co2_equivalent' : "other"
})
gas_df
```

```python
gas_df.columns = ["area","year","value","category"]
gas_df
```

```python
countries = duplicateEntries(list(gas_df["area"].unique()),len(gas_df["year"].unique()))
years = duplicateList(list(gas_df["year"].unique()),len(gas_df["area"].unique()))
columns = listAppender(gas_df.columns.tolist()[0:2],gas_df.category.unique().tolist())
values = valSplit(gas_df)
data = dataPrepper(countries,years,values)
```

```python
new_df = pd.DataFrame(data = data, columns = columns)
new_df
```

```python
# Save data to CSV
file_path = "Resources/Greenhouse_data_cleaned.csv"
new_df.to_csv(file_path, index=False)
```
